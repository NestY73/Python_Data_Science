Задание 2

```python
#Создаем массив из 51 числа в диапазоне от 0 до 10
t = np.linspace(0,10,51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t,f, label = 'cos(t)', color='green')
plt.show()
```

```python
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#808080', 'family': 'Arial'}
label_dict = {'fontsize': '14', 'color': '#808080', 'family': 'Arial'}
plt.plot(t,f, label = 'cos(t)', color='green')


plt.show()
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t,f, label = 'cos(t)', color='green')
plt.title('Значения f(t)', fontdict=title_dict)
plt.xlabel('Значения f', fontdict=label_dict)
plt.ylabel('Значения t', fontdict=label_dict)

plt.legend(labels=['cos(t)'])
plt.show()
```

Задание 3

```python
x = np.linspace(-3,3,51)
y1 = x**2
y2 = 2*x+0.5
y3 = -3*x-1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('График y4')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
